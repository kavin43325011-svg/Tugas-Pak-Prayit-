# **PRATIKUM 10**

**PRAKTIKUM 1 buble short**

```python
def bubble_sort(data):
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            if data[j] > data[j + 1]:
                data[j], data[j + 1] = data[j + 1], data[j]
    return data

angka = [64, 34, 25, 12, 22, 11, 90]
print("Data sebelum diurutkan:", angka)
print("Data setelah Bubble Sort:", bubble_sort(angka))
```

**PRAKTIKUM 2 Selection Sort**

```python
def selection_sort(data):
    n = len(data)
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            if data[j] < data[min_index]:
                min_index = j
        data[i], data[min_index] = data[min_index], data[i]
    return data

angka = [64, 25, 12, 22, 11]
print("Data sebelum diurutkan:", angka)
print("Data setelah Selection Sort:", selection_sort(angka))
```

**PRAKTIKUM 3**

```python
def insertion_sort(data):
    n = len(data)
    for i in range(1, n):
        key = data[i]
        j = i - 1
        while j >= 0 and data[j] > key:
            data[j + 1] = data[j]
            j -= 1
        data[j + 1] = key
    return data

angka = [12, 11, 13, 5, 6]
print("Data sebelum diurutkan:", angka)
print("Data setelah Insertion Sort:", insertion_sort(angka))
```

**PRAKTIKUM 4 Studi Kasus: Mengurutkan Nilai Mahasiswa**

```python
def bubble_sort_nilai(data):
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            if data[j][1] < data[j + 1][1]:
                data[j], data[j + 1] = data[j + 1], data[j]
    return data

mahasiswa = [("Andi", 80), ("Budi", 70), ("Cici", 90), ("Doni", 85)]
print("Data sebelum diurutkan:", mahasiswa)
print("Data setelah diurutkan (desc):", bubble_sort_nilai(mahasiswa))
```

**PRAKTIKUM 5 Perbandingan Waktu Eksekusi Sorting**

```python
import random, time

def bubble_sort(data):
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            if data[j] > data[j + 1]:
                data[j], data[j + 1] = data[j + 1], data[j]
    return data

def selection_sort(data):
    n = len(data)
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            if data[j] < data[min_index]:
                min_index = j
        data[i], data[min_index] = data[min_index], data[i]
    return data

def insertion_sort(data):
    n = len(data)
    for i in range(1, n):
        key = data[i]
        j = i - 1
        while j >= 0 and data[j] > key:
            data[j + 1] = data[j]
            j -= 1
        data[j + 1] = key
    return data

data = [random.randint(1, 1000) for _ in range(500)]
fungsi_sort = [bubble_sort, selection_sort, insertion_sort]

for func in fungsi_sort:
    salinan = data.copy()
    start = time.time()
    func(salinan)
    end = time.time()
    print(f"{func.__name__} membutuhkan waktu {end - start:.5f} detik")
```

**Praktikum 06 â€“ Sorting Data Barang (Nama & Harga)**

```python
def bubble_sort_key(data, key_index=0, reverse=False):
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            if reverse:
                if data[j][key_index] < data[j + 1][key_index]:
                    data[j], data[j + 1] = data[j + 1], data[j]
            else:
                if data[j][key_index] > data[j + 1][key_index]:
                    data[j], data[j + 1] = data[j + 1], data[j]
    return data

barang = [("Buku", 15000), ("Pensil", 5000), ("Tas", 75000), ("Pulpen", 10000)]
print("Urut berdasarkan nama (asc):", bubble_sort_key(barang.copy(), 0))
print("Urut berdasarkan harga (desc):", bubble_sort_key(barang.copy(), 1, reverse=True))
```

```python
def bubble_sort_desc(data):
    n = len(data)
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            if data[j] < data[j + 1]:
                data[j], data[j + 1] = data[j + 1], data[j]
                swapped = True
        if not swapped:
            break
    return data

angka = [64, 34, 25, 12, 22, 11, 90]
print(f"{bubble_sort_desc(angka)}")
```

**TUGAS 5**

```python
import random

def count_bubble_sort(data):
    n = len(data)
    perbandingan = 0
    for i in range(n):
        for j in range(0, n - i - 1):
            perbandingan += 1
            if data[j] > data[j + 1]:
                data[j], data[j + 1] = data[j + 1], data[j]
    return perbandingan

def count_selection_sort(data):
    n = len(data)
    perbandingan = 0
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            perbandingan += 1
            if data[j] < data[min_index]:
                min_index = j
        data[i], data[min_index] = data[min_index], data[i]
    return perbandingan

def count_insertion_sort(data):
    n = len(data)
    perbandingan = 0
    for i in range(1, n):
        key = data[i]
        j = i - 1
        while j >= 0:
            perbandingan += 1
            if data[j] > key:
                data[j + 1] = data[j]
                j -= 1
            else:
                break
        data[j + 1] = key
    return perbandingan

print("Analisis Kompleksitas:")
print("{:<8} | {:<8} | {:<8} | {:<8}".format("N", "Bubble", "Selection", "Insertion"))
print("-" * 38)

sizes = [10, 50, 100]

for n in sizes:
    data_random = [random.randint(1, 1000) for _ in range(n)]

    b_comp = count_bubble_sort(data_random.copy())
    s_comp = count_selection_sort(data_random.copy())
    i_comp = count_insertion_sort(data_random.copy())

    print("{:<8} | {:<8} | {:<8} | {:<8}".format(n, b_comp, s_comp, i_comp))
```

**TUGAS 6**

```python
def sort_by_choice(data, key_index=0, reverse=False):
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            val1 = data[j][key_index]
            val2 = data[j + 1][key_index]

            should_swap = False
            if reverse:
                if val1 < val2:
                    should_swap = True
            else:
                if val1 > val2:
                    should_swap = True

            if should_swap:
                data[j], data[j + 1] = data[j + 1], data[j]
    return data

barang = [
    ("Pensil", 5000),
    ("Tas", 75000),
    ("Buku", 15000),
    ("Pulpen", 10000)
]

def proyek_mini_manajemen_barang():
    print("\nProyek Mini: Manajemen Barang")
    print("--------------------------------")
    print(f"Data Awal: {barang}")

    while True:
        key_input = input("Urutkan berdasarkan (1: Nama, 2: Harga): ")
        if key_input == '1':
            key_index = 0
            break
        elif key_input == '2':
            key_index = 1
            break
        else:
            print("Pilihan tidak valid.")

    while True:
        direction_input = input("Arah urutan (A: Ascending, D: Descending): ")
        if direction_input.upper() == 'A':
            reverse = False
            break
        elif direction_input.upper() == 'D':
            reverse = True
            break
        else:
            print("Pilihan tidak valid.")

    data_terurut = sort_by_choice(barang.copy(), key_index, reverse)

    print("\n--- Hasil ---")
    print(data_terurut)
    print("-------------")

proyek_mini_manajemen_barang()
```
