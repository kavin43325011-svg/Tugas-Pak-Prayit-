# **PRATIKUM 11**

**Praktikum 01 – Sequential Search pada Data Angka**

```python
def sequential_search(data, target):
    for i in range(len(data)):
        if data[i] == target:
            return f"Data {target} ditemukan di indeks {i}"
    return f"Data {target} tidak ditemukan"

angka = [10, 20, 30, 40, 50]
print("Data:", angka)
print(sequential_search(angka, 30))
print(sequential_search(angka, 100))
```

**Praktikum 02 – Binary Search pada Data Angka**

```python
def binary_search(data, target):
    left, right = 0, len(data)-1
    while left <= right:
        mid = (left + right) // 2
        if data[mid] == target:
            return f"Data {target} ditemukan di indeks {mid}"
        elif data[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return f"Data {target} tidak ditemukan"

angka = [10, 20, 30, 40, 50]
print("Data:", angka)
print(binary_search(angka, 40))
print(binary_search(angka, 100))
```

**PRAKTIKUM 3 Sequential Search untuk Data Nama**

```python
def sequential_search_nama(data, target):
    for i in range(len(data)):
        if data[i].lower() == target.lower():
            return f"Nama {target} ditemukan di indeks {i}"
    return f"Nama {target} tidak ditemukan"

nama = ["Andi", "Budi", "Cici", "Doni"]
print("Data:", nama)
print(sequential_search_nama(nama, "Cici"))
print(sequential_search_nama(nama, "Eka"))
```

**PRAKTIKUM 4 Studi Kasus Nilai Mahasiswa dengan Binary Search Mahasiswa**

**PRAKTIKUM 5 Perbandingan Waktu Sequential vs Binary Search**

```python
import time

def sequential_search(data, target):
    for i in range(len(data)):
        if data[i] == target:
            return i
    return -1

def binary_search(data, target):
    left, right = 0, len(data)-1
    while left <= right:
        mid = (left + right) // 2
        if data[mid] == target:
            return mid
        elif data[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1

data = list(range(1, 100001))
target = 99999

start = time.time()
sequential_search(data, target)
end = time.time()
print("Sequential Search waktu:", end - start, "detik")

start = time.time()
binary_search(data, target)
end = time.time()
print("Binary Search waktu:", end - start, "detik")
```

**PRAKTIKUM 6 Kombinasi Sorting dan Searching**

```python
import random

def insertion_sort(data):
    for i in range(1, len(data)):
        key = data[i]
        j = i - 1
        while j >= 0 and data[j] > key:
            data[j + 1] = data[j]
            j -= 1
        data[j + 1] = key
    return data

def binary_search(data, target):
    left, right = 0, len(data)-1
    while left <= right:
        mid = (left + right) // 2
        if data[mid] == target:
            return f"{target} ditemukan di indeks {mid}"
        elif data[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return f"{target} tidak ditemukan"

data = [random.randint(1, 100) for _ in range(15)]
print("Data acak sebelum diurutkan:", data)
urut = insertion_sort(data)
print("Data setelah diurutkan:", urut)
target = urut[7]  # ambil angka ke-8 sebagai target
print(binary_search(urut, target))
```

# **PENUGASAN**

**TUGAS 1 Variasi Sequential Search**

```python
def sequential_search_count(data, target):
    comparisons = 0
    for i in range(len(data)):
        comparisons += 1
        if data[i] == target:
            return f"Data {target} ditemukan di indeks {i} dengan {comparisons} perbandingan."
    return f"Data {target} tidak ditemukan setelah {comparisons} perbandingan."

data_seq = [10, 5, 20, 8, 30, 15]
print("Data:", data_seq)
print(sequential_search_count(data_seq, 30))
print(sequential_search_count(data_seq, 99))
```

**TUGAS 2 Variasi Binary Search**

```python
def binary_search_steps(data, target):
    left, right = 0, len(data) - 1
    steps = 0
    while left <= right:
        steps += 1
        mid = (left + right) // 2
        if data[mid] == target:
            return f"Data {target} ditemukan di indeks {mid} dalam {steps} langkah (iterasi)."
        elif data[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return f"Data {target} tidak ditemukan setelah {steps} langkah (iterasi)."

data_bin = [10, 20, 30, 40, 50, 60, 70, 80]
print("Data terurut:", data_bin)
print(binary_search_steps(data_bin, 70))
print(binary_search_steps(data_bin, 25))
```

**TUGAS 3 Studi Kasus Nama Mahasiswa**

```python
def sequential_search_mahasiswa(data, target):
    for i in range(len(data)):
        if data[i].lower() == target.lower():
            return f"Nama '{target}' ditemukan di indeks {i}."
    return f"Nama '{target}' tidak ditemukan dalam daftar."

def studi_kasus_mahasiswa():
    nama_mahasiswa = []
    print("Masukkan daftar nama mahasiswa (ketik 'selesai' untuk mengakhiri input):")
    while True:
        nama = input("> ")
        if nama.lower() == 'selesai':
            break
        nama_mahasiswa.append(nama)

    if not nama_mahasiswa:
        print("Daftar nama kosong.")
        return

    print("\nDaftar nama yang dimasukkan:", nama_mahasiswa)
    target_nama = input("Masukkan nama mahasiswa yang ingin dicari: ")

    hasil = sequential_search_mahasiswa(nama_mahasiswa, target_nama)
    print(hasil)

studi_kasus_mahasiswa()
```

**TUGAS 4 Studi Kasus Barang Toko**

```python
def binary_search_harga(data_list, target_harga):
    left, right = 0, len(data_list) - 1
    while left <= right:
        mid = (left + right) // 2
        harga_mid = data_list[mid][1]

        if harga_mid == target_harga:
            found_items = []
            # Cari semua barang dengan harga yang sama
            k = mid
            while k >= 0 and data_list[k][1] == target_harga:
                found_items.append(data_list[k])
                k -= 1
            k = mid + 1
            while k < len(data_list) and data_list[k][1] == target_harga:
                found_items.append(data_list[k])
                k += 1

            return f"Ditemukan {len(found_items)} barang dengan harga {target_harga}: {found_items}"
        elif harga_mid < target_harga:
            left = mid + 1
        else:
            right = mid - 1
    return f"Barang dengan harga {target_harga} tidak ditemukan."

daftar_barang = [
    ("Pensil", 5000),
    ("Buku Tulis", 15000),
    ("Penghapus", 2000),
    ("Stabilo", 10000),
    ("Kertas HVS", 15000)
]

# Urutkan berdasarkan harga (indeks 1)
daftar_barang_urut = sorted(daftar_barang, key=lambda x: x[1])
print("Daftar barang urut berdasarkan harga:", daftar_barang_urut)

harga_cari = 15000
print(binary_search_harga(daftar_barang_urut, harga_cari))
harga_cari_lain = 7500
print(binary_search_harga(daftar_barang_urut, harga_cari_lain))
```

**TUGAS 5 Analisis Kompleksitas**

```python
import time
import random

def sequential_search_only(data, target):
    for i in range(len(data)):
        if data[i] == target:
            return i
    return -1

def binary_search_only(data, target):
    left, right = 0, len(data) - 1
    while left <= right:
        mid = (left + right) // 2
        if data[mid] == target:
            return mid
        elif data[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1

def run_analysis(sizes):
    results = []
    for size in sizes:
        data = list(range(1, size + 1))
        # Target di akhir (kasus terburuk Sequential)
        target = size - 1

        # Sequential Search
        start = time.time()
        sequential_search_only(data, target)
        time_seq = time.time() - start

        # Binary Search
        start = time.time()
        binary_search_only(data, target)
        time_bin = time.time() - start

        results.append((size, time_seq, time_bin))

    print("----------------------------------------------------------")
    print("| Ukuran Data (N) | Sequential Search (detik) | Binary Search (detik) |")
    print("----------------------------------------------------------")
    for size, t_seq, t_bin in results:
        print(f"| {size:<15} | {t_seq:<25.10f} | {t_bin:<21.10f} |")
    print("----------------------------------------------------------")

run_analysis([100, 1000, 10000])
```

```python
def binary_search_mahasiswa(data_list, target_nilai):
    # data_list sudah terurut berdasarkan nilai
    left, right = 0, len(data_list) - 1
    results = []

    while left <= right:
        mid = (left + right) // 2
        nilai_mid = data_list[mid][1]

        if nilai_mid == target_nilai:
            # Ditemukan, cari semua entri dengan nilai yang sama
            k = mid
            while k >= 0 and data_list[k][1] == target_nilai:
                results.append(data_list[k])
                k -= 1
            k = mid + 1
            while k < len(data_list) and data_list[k][1] == target_nilai:
                results.append(data_list[k])
                k += 1

            # Hapus duplikat dari pencarian dua arah
            unique_results = []
            for item in results:
                if item not in unique_results:
                    unique_results.append(item)

            return f"Ditemukan {len(unique_results)} mahasiswa dengan nilai {target_nilai}: {unique_results}"

        elif nilai_mid < target_nilai:
            left = mid + 1
        else:
            right = mid - 1

    return f"Tidak ada mahasiswa dengan nilai {target_nilai} dalam daftar."

def proyek_mini_nilai():
    data_mahasiswa = [
        ("Andi", 85),
        ("Budi", 70),
        ("Cici", 90),
        ("Doni", 70),
        ("Eka", 85),
        ("Fara", 95)
    ]

    # Urutkan data berdasarkan nilai (indeks 1)
    data_terurut = sorted(data_mahasiswa, key=lambda x: x[1])

    print("Daftar Mahasiswa (Nama, Nilai) terurut:", data_terurut)

    try:
        nilai_cari = int(input("Masukkan nilai mahasiswa yang ingin dicari (contoh: 85): "))
        hasil = binary_search_mahasiswa(data_terurut, nilai_cari)
        print(hasil)
    except ValueError:
        print("Input nilai tidak valid. Harap masukkan angka.")

proyek_mini_nilai()
```
