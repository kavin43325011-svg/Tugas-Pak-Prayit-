**PRATIKUM 7**

```python
def decorator_sapa(fungsi):
    def wrapper():
        print("Sebelum fungsi dijalankan...")
        fungsi()
        print("Setelah fungsi dijalankan...")
    return wrapper

@decorator_sapa
def sapa():
    print("Halo, Mahasiswa Polines!")

sapa()
```

**TUGAS 8**

```python
def logger(fungsi):
    def wrapper():
        print("Memulai fungsi...")
        fungsi()
        print("Fungsi selesai dijalankan.")
    return wrapper

@logger
def halo():
    print("Halo, mahasiswa Polines!")

halo()
```

**TUGAS 9**

```python
# a. fungsi input_nilai()
def input_nilai():
    return [float(x) for x in input("Masukkan nilai mahasiswa (pisahkan dengan spasi): ").split()]

# b. fungsi hitung_rata()
def hitung_rata(nilai):
    return sum(nilai) / len(nilai)

# c. fungsi status_kelulusan()
def status_kelulusan(rata):
    return "Lulus" if rata >= 60 else "Tidak Lulus"

# d. lambda untuk nilai akhir
nilai_akhir = lambda tugas, ujian: (0.4 * tugas) + (0.6 * ujian)

# Program utama
nilai = input_nilai()
rata = hitung_rata(nilai)
print("Rata-rata:", rata)
print("Status:", status_kelulusan(rata))

tugas = float(input("Nilai tugas: "))
ujian = float(input("Nilai ujian: "))
print("Nilai akhir:", nilai_akhir(tugas, ujian))
```
