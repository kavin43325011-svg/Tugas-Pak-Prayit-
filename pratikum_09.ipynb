```python
import math

def luas_lingkaran(r):
    return math.pi * (r ** 2)

print("Luas lingkaran jari-jari 7 =", luas_lingkaran(7))
print("Luas lingkaran jari-jari 10 =", luas_lingkaran(10))
```

**TUGAS 4**

```python
%%writefile konversi.py
def c_to_f(angka):
  f = angka * 1.8 + 32
  print(f"Konversi dari {angka} Celcius ke Fahrenheit adalah: °{f}")

def f_to_c(angka):
  c = (angka - 32 ) / 1.8
  print(f"Konversi dari {angka} Fahrenheit ke Celcius adalah: °{c}")
```

```python
import konversi

print("----- Konversi -----")
print("1. Celcius ke Fahrenheit")
print("2. Fahrenheit ke Celcius")
print("--------------------")
print()
pilih = int(input("Masukkan angka: "))


if pilih == 1:
  angka =int(input("Masukkan angka yang ingin anda konversi menjadi Fahrenheit: "))
  konversi.c_to_f(angka)
else:
  angka =int(input("Masukkan angka yang ingin anda konversi menjadi Celcius: "))
  konversi.f_to_c(angka)
```
